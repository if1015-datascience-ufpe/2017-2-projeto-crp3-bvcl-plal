```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
season_2014_15 = pd.read_csv('season_players/2014-2015.csv')
season_2014_15 = season_2014_15.drop('Unnamed: 0', axis=1)

season_2015_16 = pd.read_csv('season_players/2015-2016.csv')
season_2015_16 = season_2015_16.drop('Unnamed: 0', axis=1)

season_2016_17 = pd.read_csv('season_players/2016-2017.csv')
season_2016_17 = season_2016_17.drop('Unnamed: 0', axis=1)

season_2017_18 = pd.read_csv('season_players/2017-2018.csv')
season_2017_18 = season_2017_18.drop('Unnamed: 0', axis=1)

```

```python
df_final_full = pd.concat([season_2014_15,season_2015_16,season_2016_17,season_2017_18], ignore_index=True)
df_final = df_final_full.drop(['Team','Season','Name'],axis=1).fillna(0)
df_final
```

```python
model = KMeans(n_clusters = 100, random_state = 7)
model.fit(df_final)
clusters = model.labels_
print(len(clusters))
```

```python
df_final_full['Cluster'] = clusters
df_final_full[df_final_full['Name'] == 'LeBron James']
```

Partindo da premissa que LeBron James é um jogador de sucesso, podemos buscar respostas para a seguinte pergunta: quem também é considerado um jogador de sucesso? Isso pode ser respondido, analisando quais jogadores estão no mesmo cluster no mesmo período em que LeBron James esteve no ápice de sua carreira: 2015-2016. Logo, o cluster a ser analisado é o 29.

```python
df_final_full[df_final_full['Cluster'] == 29]
```

Vemos, a partir da filtragem realizada, os jogadores que podem ser considerados "bem sucedidos", ao pertencer ao cluster de LeBron James no ápice de sua carreira. É notável a presença de nomes já conhecidos na NBA, como Carmelo Anthony e Rudy Gay, que já são conhecidos na NBA como bons jogadores, o que torna coerente sua presença no mesmo grupo que LeBron. Algo relevante a ser notado também, é que alguns jogadores, como Kawhi Leonard e Victor Oladipo, no início da carreira, apresentaram performances similares às de LeBron, o que não se repetiu nos anos seguintes, já que caso contrário estariam no mesmo grupo que LeBron.
