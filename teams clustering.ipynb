```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
season_2014_15 = pd.read_csv('season_stats/2014-2015_season_stats.csv')
season_2014_15 = season_2014_15.drop('Unnamed: 0', axis=1)

season_2015_16 = pd.read_csv('season_stats/2015-2016_season_stats.csv')
season_2015_16 = season_2015_16.drop('Unnamed: 0', axis=1)

season_2016_17 = pd.read_csv('season_stats/2016-2017_season_stats.csv')
season_2016_17 = season_2016_17.drop('Unnamed: 0', axis=1)

season_2017_18 = pd.read_csv('season_stats/2017-2018_season_stats.csv')
season_2017_18 = season_2017_18.drop('Unnamed: 0', axis=1)

df_final_full = pd.concat([season_2014_15,season_2015_16,season_2016_17,season_2017_18], ignore_index = True)
df_final = df_final_full.drop(['Season','CONF','Team'], axis=1).fillna(0)
df_final
```

```python
model = KMeans(n_clusters = 15, random_state = 7)
model.fit(df_final)
clusters = model.labels_
print(clusters)
```

```python
df_final_full['Cluster'] = clusters
df_final_full[df_final_full['Team'] == 'C']
```
