```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests as rq
```

```python
url = 'https://www.basketball-reference.com/'
req = rq.get(url)
page = BeautifulSoup(req.text, 'html.parser')
print(page)
```

```python
data = []
for a in page.find_all("tr", class_="full_table"):
    for b in a.find_all("a"):
        if('20' in b["href"]):
            data.append(dict(zip(['team','link'], [b["title"], "http://www.basketball-reference.com" + b["href"]])))
df = pd.DataFrame(data, columns=['team','link'])
df
```

```python
all_links = []
for row in df.itertuples():
    team = row[1]
    link = row[2]
    l = link[0:link.rfind('/')+1]
    for i in range(2008,2019):
        all_links.append(dict(zip(["team","links"],[team,l + str(i) + ".html"])))
#print(all_links)
tets = pd.DataFrame(all_links, columns=["team","links"])
tets
```

```python
def getRoster(link):
    url = link
    req = rq.get(url)
    team_page = BeautifulSoup(req.text, 'html.parser')
    i=0
    aux=[]
    name="";
    ht="";
    wt="";
    bday="";
    exp="";
    college=""
    for t in team_page.find_all('table',id="roster"):
        for a in t.find('tbody').find_all('td'):
            if(i==0):
                name=a.text;
            elif(i==2):
                ht=a.text;
            elif(i==3):
                wt=a.text;
            elif(i==4):
                bday=a.text;
            elif(i==6):
                if("R" in a.text):
                    exp=0;
                else:
                    exp=a.text;
            elif(i==7):
                college=a.text;
            i=i+1
            if(i==8):
                i=0;
                aux.append(dict(zip(["name","height","weight","bday","exp","college"],[name,ht,wt,bday,exp,college])))
                name="";
                ht="";
                wt="";
                bday="";
                exp="";
                college="";
    return pd.DataFrame(aux, columns=["name","height","weight","bday","exp","college"])
#df = getRoster('https://www.basketball-reference.com/teams/DET/2018.html')
#df
```
