```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup, Comment
import requests as rq
```

```python
url = 'https://www.basketball-reference.com/'
req = rq.get(url)
page = BeautifulSoup(req.text, 'html.parser')
#print(page)
```

```python
data = []
for a in page.find_all("tr", class_="full_table"):
    for b in a.find_all("a"):
        if('20' in b["href"]):
            data.append(dict(zip(['team','link'], [b["title"], "http://www.basketball-reference.com" + b["href"]])))
df = pd.DataFrame(data, columns=['team','link'])
df
```

```python
all_links = []
for row in df.itertuples():
    team = row[1]
    link = row[2]
    l = link[0:link.rfind('/')+1]
    for i in range(2008,2019):
        all_links.append(dict(zip(["team","links"],[team,l + str(i) + ".html"])))
#print(all_links)
tets = pd.DataFrame(all_links, columns=["team","links"])
tets
```
