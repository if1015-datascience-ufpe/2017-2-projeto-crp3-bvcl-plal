```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup, Comment
import requests as rq
```

```python
url = 'https://www.basketball-reference.com/'
req = rq.get(url)
page = BeautifulSoup(req.text, 'html.parser')
#print(page)
```

```python
data = []
for a in page.find_all("tr", class_="full_table"):
    for b in a.find_all("a"):
        if('20' in b["href"]):
            data.append(dict(zip(['team','link'], [b["title"], "http://www.basketball-reference.com" + b["href"]])))
df = pd.DataFrame(data, columns=['team','link'])
df
```

```python
all_links = []
for row in df.itertuples():
    team = row[1]
    link = row[2]
    l = link[0:link.rfind('/')+1]
    for i in range(2015,2019):
        all_links.append(dict(zip(["team","links"],[team,l + str(i) + ".html"])))
#print(all_links)
df_links = pd.DataFrame(all_links, columns=["team","links"])
df_links
```

```python
def find_comment_index(table,comments):
    index = 0
    for comment in comments:
        if table in comment:
            return index
        index += 1
```

```python
stats_labels = ['MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','FT','FTA','FT%',
                'ORD','DRB','TRB','AST','STL','BLK','TOV','PF','PTS']

def get_season_per_game_data(link):

    req = rq.get(link)
    page = BeautifulSoup(req.text, 'html.parser')

    comments=page.find_all(string=lambda text:isinstance(text,Comment))
    comm_index = find_comment_index("team_and_opponent",comments)
    #print(comm_index)
    team_opp = BeautifulSoup(comments[comm_index], 'html.parser')
    team_opp
    table_team_opp = team_opp.find("table",id="team_and_opponent")
    column = 0
    data = []
    for r in table_team_opp.find_all('td'):
        if(column >= 24 and column <= 45):
            data.append(float(r.text))
        column += 1
    per_game_data = dict(zip(stats_labels,data))
    return per_game_data

#data_aux = get_season_per_game_data('https://www.basketball-reference.com/teams/BOS/2013.html')


def get_team_stats(team):
    
    data_ = []
    for row in df_links.itertuples():
        if row[1] == team:
            data_.append(get_season_per_game_data(row[2]))
    df_team = pd.DataFrame(data_)
    cols = df_team.columns.tolist()
    cols = stats_labels
    df_team = df_team[cols]
    return df_team

get_team_stats('Minnesota Timberwolves')
```

```python
import time

teams = df_links['team'].tolist()
teams_list = list(set(teams))

for team in teams_list:
    print(team)
    time.sleep(0.01)
    df_team = get_team_stats(team)
    team_name = team.replace(" ","_")
    df_team.to_csv('team_stats/' + team_name + '_stats_2014_to_2018.csv')
```

```python
def get_roster_info(link):
    url = link
    req = rq.get(url)
    team_page = BeautifulSoup(req.text, 'html.parser')
    i=0
    aux=[]
    name="";
    ht="";
    wt="";
    bday="";
    exp="";
    college="";
    for t in team_page.find_all('table',id="roster"):
        for a in t.find('tbody').find_all('td'):
            if(i==0):
                name=a.text;
            elif(i==2):
                ht=a.text;
            elif(i==3):
                wt=a.text;
            elif(i==4):
                bday=a.text;
            elif(i==6):
                if("R" in a.text):
                    exp=0;
                else:
                    exp=a.text;
            elif(i==7):
                college=a.text;
            i=i+1
            if(i==8):
                i=0;
                aux.append(dict(zip(["name","height","weight","bday","exp","college"],[name,ht,wt,bday,exp,college])))
                name="";
                ht="";
                wt="";
                bday="";
                exp="";
                college="";
    return pd.DataFrame(aux, columns=["name","height","weight","bday","exp","college"])
```

```python
links = df_links['links'].tolist()

year = 2014
for a in df_links.itertuples():
    team_name = a[1].replace(" ","_")
    df_roster = get_roster_info(a[2])
    df_roster.to_csv('roster_info/' + team_name + '/roster_info_' + str(year) + '-' + str(year+1) + '.csv')
    year += 1
    if year == 2018:
        year = 2014
    
```

```python
roster_stats_labels = ['Name','Age','G','GS','MP','FG','FGA','FG%','3P','3PA','3P%','2P','2PA','2P%','eFG%',
                       'FT','FTA','FT%','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS/G']

def get_roster_stats(link):
    
    req = rq.get(link)
    page = BeautifulSoup(req.text, 'html.parser')

    comments=page.find_all(string=lambda text:isinstance(text,Comment))
    comm_index = find_comment_index("Per Game Table",comments)
    #print(comm_index)
    per_game_stats = BeautifulSoup(comments[comm_index], 'html.parser')
    table_per_game_stats = per_game_stats.find("table",id='per_game')
    index = 0
    data = []
    row = []
    for r in table_per_game_stats.find_all('td'):
        row.append(r.text)
        index += 1
        if index % 27 == 0:
            data.append(dict(zip(roster_stats_labels,row)))
            row = []
    df_roster_stats = pd.DataFrame(data)
    cols = df_roster_stats.columns.tolist()
    cols = roster_stats_labels
    df_roster_stats = df_roster_stats[cols]
    return df_roster_stats

get_roster_stats('https://www.basketball-reference.com/teams/CHI/2018.html')
```

```python
year = 2014
for a in df_links.itertuples():
    team_name = a[1].replace(" ","_")
    df_roster_stats = get_roster_stats(a[2])
    df_roster_stats.to_csv('roster_stats/' + team_name + '/roster_stats_' + str(year) + '-' + str(year+1) + '.csv')
    year += 1
    if year == 2018:
        year = 2014
```
